```python
# mengimport dataset gambar
import zipfile
#path ke zip file
zip_file_path = '/content/drive/MyDrive/imagess.zip'

# extract ke direktori images
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

picture_directory = '/content/imagess/Train'

# mengambil list gambar dan tidak termasuk xml
image_files = [filename for filename in os.listdir(picture_directory) if not filename.endswith('.xml')]

# mengurutkan gambar
image_files.sort()

# ambil 8 gambar pertama
image_files = image_files[:8]

# inisialisasi  subplot
fig, axs = plt.subplots(2, 4, figsize=(12, 6))
fig.suptitle("Contoh dataset gambar yang digunakan")

# melakukan iterasi direktori tempat gambar
for i, filename in enumerate(image_files):
    try:
        img = Image.open(os.path.join(picture_directory, filename))
        axs[i // 4, i % 4].imshow(img)
        axs[i // 4, i % 4].axis('off')
        # Smenampilkan nama gambar
        axs[i // 4, i % 4].set_title(filename)
    except Exception as e:
        print(f"Can't read file {os.path.join(picture_directory, filename)}: {e}")

# tampilkan gambar
plt.tight_layout()
plt.show()
```
